# ML

```python
import pandas as pd
import seaborn as sns
from scipy import stats

url = 'https://docs.google.com/spreadsheets/d/1sNEhLYCBXqrjaVBGwpxNdKSJuK4TI3K38G8-jOm04z0/edit?usp=sharing'
url = url[:url.find('/edit')] + '/export?format=csv'

df = pd.read_csv(url)
```

```python
df = df.rename(columns={
    'TV': 'tv',
    'Billboards': 'bil',
    'Google_Ads': 'gads',
    'Social_Media': 'soc',
    'Influencer_Marketing': 'inf',
    'Affiliate_Marketing': 'aff',
    'Product_Sold': 'sold'
}, inplace=False)
```

```python
df
```

```python
sns.pairplot(df, kind='reg')
```

```python
sns.jointplot(
    df,
    x='aff',
    y='sold',
    kind='kde'
)
```

```python
from sklearn.model_selection import train_test_split
target = 'sold'
X = df.drop(columns=[target])
y = df[target]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
```

```python
model_lr = LinearRegression()
model_lr.fit(x_train, y_train)
```

```python
X.columns
```

```python
model_lr.coef_
```

```python
pred_train = model_lr.predict(x_train)
pred_test = model_lr.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_train, pred_train)), np.sqrt(mean_squared_error(y_test, pred_test))
```

```python
mean_absolute_percentage_error(y_train, pred_train), mean_absolute_percentage_error(y_test, pred_test)
```

```python
sns.scatterplot(x=y_test, y=pred_test)
```

```python
sns.scatterplot(x=y_test, y=pred_test-y_test)
sns.lineplot(x=[2000, 11000], y=[0, 0], color='green')
```

```python
sns.histplot(pred_test-y_test)
```

```python
# Fork X <- Z -> Y
n = 1000
z = np.random.binomial(1, 0.5, n)
x = np.random.binomial(1, (1-z)*0.1 + z*0.9, n)
y = np.random.binomial(1, (1-z)*0.1 + z*0.9, n)
```

```python
df = pd.DataFrame({'x': x, 'y': y, 'z': z})
```

```python
df.corr()
```

```python
df[df.z == 1].corr()
```

```python
df[df.z == 0].corr()
```

```python
# Pipe X -> Z -> Y
n = 1000
x = np.random.binomial(1, 0.5, n)
z = np.random.binomial(1, (1-x)*0.1 + x*0.9, n)
y = np.random.binomial(1, (1-z)*0.1 + z*0.9, n)
```

```python
df = pd.DataFrame({'x': x, 'y': y, 'z': z})
```

```python
df.corr()
```

```python
df[df.z == 1].corr()
```

```python
df[df.z == 0].corr()
```

```python
# Collider X -> Z <- Y
n = 1000
x = np.random.binomial(1, 0.5, n)
y = np.random.binomial(1, 0.5, n)
z = [np.random.binomial(1, 0.9 if i > 0 else 0.2) for i in x+y]
```

```python
df = pd.DataFrame({'x': x, 'y': y, 'z': z})
```

```python
df.corr()
```

```python
df[df.z == 1].corr()
```

```python
df[df.z == 0].corr()
```
