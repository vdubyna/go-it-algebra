## Завдання 1

В партії з 50 деталей 5 нестандартних. Визнач ймовірність того, що серед обраних навмання для перевірки шести деталей дві виявляться нестандартними. Розв'яжи аналітично.

Примітка: в даній задачі не потрібно рахувати конкретне число.

---
Розв'язання:

Ймовірність — це відношення цієї кількості сприятливих варіантів до загальної кількості всіх можливих варіантів вибору 6 деталей із 50.
Ми шукаємо ситуацію, коли одночасно виконуються дві умови:
	•	Вибрані дві нестандартні деталі.
	•	Вибрані чотири стандартні деталі.
Для цього використовуєм комбінацію

$P = \frac{\text{Кількість сприятливих випадків}}{\text{Загальна кількість можливих випадків}} = \frac{C(5, 2) \cdot C(45, 4)}{C(50, 6)}$ 

$C(5, 2) = \frac{5!}{2!(5-2)!} = \frac{5 \cdot 4}{2 \cdot 1} = 10$

$C(45, 4) = \frac{45!}{4!(45-4)!} = \frac{45 \cdot 44 \cdot 43 \cdot 42}{4 \cdot 3 \cdot 2 \cdot 1} = 148,995$

$C(50, 6) = \frac{50!}{6!(50-6)!} = \frac{50 \cdot 49 \cdot 48 \cdot 47 \cdot 46 \cdot 45}{6 \cdot 5 \cdot 4 \cdot 3 \cdot 2 \cdot 1} = 15,890,700$

$P = \frac{10 \cdot 148,995}{15,890,700} = \frac{1,489,950}{15,890,700} \approx 0.0938$

Отже, ймовірність дорівнює 9.38%

## Завдання 2

В тирі стрілець проводить 7 пострілів по мішені з ймовірністю влучання кожного 0.8. Яка ймовірність того, що буде не менше за 5 влучань? Розв'яжи аналітично.

---
Розв'язання:
Ця задача стосується біноміального розподілу
Ймовірність того, що стрілець влучить `k`  разів у `n` пострілах визначається біноміальним розподілом:

$P(X = k) = C(n, k) \cdot p^k \cdot (1 - p)^{n - k}$

де:
* $C(n, k) = \frac{n!}{k!(n - k)!}$ — число способів вибрати $k$ влучань з $n$ пострілів;
* $p$ — ймовірність влучання;
$ $1 - p$ — ймовірність невлучання.

Потрібно знайти ймовірність $P(X \geq 5)$, тобто суму ймовірностей для 5, 6 і 7 влучань:

$P(X \geq 5) = P(X = 5) + P(X = 6) + P(X = 7)$

Обчислюємо ймовірності для $X = 5$,  $X = 6$,  $X = 7$:
Для кожного випадку скористаємося біноміальною формулою:

* $P(X = 5) = C(7, 5) \cdot 0.8^5 \cdot 0.2^2$ 
* $P(X = 6) = C(7, 6) \cdot 0.8^6 \cdot 0.2^1$ 
* $P(X = 7) = C(7, 7) \cdot 0.8^7 \cdot 0.2^0$

Для першого випадку робимо наступне:
$C(7, 5) = \frac{7!}{5! \cdot (7 - 5)!} = \frac{7 \times 6}{2 \times 1} = 21$ 
$0.8^5 = 0.32768$ 
$0.2^2 = 0.04$ 

Тепер підставляємо:

$P(X = 5) = 21 \cdot 0.32768 \cdot 0.04 = 0.2752512$

Аналогічно робимо для інших і отримуємо наступні дані:

$P(X = 6) = 7 \cdot 0.262144 \cdot 0.2 = 0.3670016$

$P(X = 7) = 1 \cdot 0.2097152 \cdot 1 = 0.2097152$

Підставляємо все у формулу суми ймовірностей:

$P(X \geq 5) = P(X = 5) + P(X = 6) + P(X = 7)$

$P(X \geq 5) = 0.2752512 + 0.3670016 + 0.2097152 = 0.851968$

Отже, ймовірність того, що стрілець влучить не менше ніж 5 разів, дорівнює 0.851968, або 85.2%.
 

## Завдання 3

Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12?  
Розв'яжи програмно, згенерувавши всі можливі комбінації.

---
Розв'язання:

````python
import itertools

# Можливі значення для кожного кубика (1 до 6)
val = [1, 2, 3, 4, 5, 6]

# Генеруємо всі можливі комбінації очок для трьох кубиків
com_set = itertools.product(val, repeat=3)

# Лічильники для сум 11 і 12
count_11 = 0
count_12 = 0

# Перевіряємо всі комбінації
for elem in com_set:
    total = sum(elem)
    if total == 11:
        count_11 += 1
    elif total == 12:
        count_12 += 1

(count_11, count_12)
````

радіуси яких 5 і 10 см відповідно. Знайди ймовірність того, що точка, кинута навмання у велике коло, потрапить також у кільце, утворене побудованими окружностями.

Розв'яжи аналітично.

---
Розв'язання:
Дано: 
* Велике коло має радіус  R = 10  см.
* Маленьке коло має радіус  r = 5  см.
* Кільце утворене між великою і маленькою окружностями.

1. Площа великого кола:
Площа великого кола з радіусом  $R = 10$  см обчислюється за формулою площі круга:

$S_{\text{велике}} = \pi R^2 = \pi \cdot 10^2 = 100\pi \, \text{см}^2$

2. Площа маленького кола:
Площа маленького кола з радіусом  $r = 5$  см:

$S_{\text{маленьке}} = \pi r^2 = \pi \cdot 5^2 = 25\pi \, \text{см}^2$

3.	Площа кільця:
Кільце — це область між великою і маленькою окружностями. Площа кільця дорівнює різниці площ великого і маленького кола:

$S_{\text{кільце}} = S_{\text{велике}} - S_{\text{маленьке}} = 100\pi - 25\pi = 75\pi \, \text{см}^2$

4.	Ймовірність попадання в кільце:
Ймовірність того, що точка, кинута випадково у велике коло, потрапить у кільце, дорівнює відношенню площі кільця до площі великого кола:

$P = \frac{S_{\text{кільце}}}{S_{\text{велике}}} = \frac{75\pi}{100\pi} = \frac{75}{100} = 0.75$

Відповідь:

Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою? Розв'яжи аналітично  

---
Розв'язання:

Нам потрібна умовна ймовірність  $P(B_1 \mid W)$  — ймовірність, що спочатку була біла куля, якщо витягнули білу. Використаємо формулу Байєса:

$P(B_1 \mid W) = \frac{P(W \mid B_1) P(B_1)}{P(W)}$

1.	Ймовірності:
* $(B_1) = 0.5 ,  P(B_2) = 0.5$ 
* $P(W \mid B_1) = 1$  (дві білі кулі)

2.	Повна ймовірність витягнути білу:

$P(W) = (1 \cdot 0.5) + (0.5 \cdot 0.5) = 0.75$

3.	Рахуємо умовну ймовірність:

$P(B_1 \mid W) = \frac{1 \cdot 0.5}{0.75} = \frac{2}{3}$

## Завдання 7. Випадкове блукання
Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. В початковий момент часу $t=0$ ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу ($t=3$).  
а) Виріши аналітично.

б) Напиши симуляцію данного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій.

---
Розв'язання:	
* Крок вгору на 2 одиниці: з імовірністю  $P(\text{вгору}) = 0.5$.
* Крок вниз на 1 одиницю: з імовірністю  $P(\text{вниз}) = 0.5$.

1.	Середня зміна ціни за один крок:
В одному кроці можливі два результати:
* Збільшення на 2 одиниці з ймовірністю 0.5.
* Зменшення на 1 одиницю з ймовірністю 0.5.
Середня зміна за один крок:

$E(\Delta X) = 0.5 \cdot 2 + 0.5 \cdot (-1) = 1 - 0.5 = 0.5$

Отже, в середньому ціна акцій зростає на 0.5 одиниці за один крок.
	
2.	Середня ціна через 3 одиниці часу:
Оскільки середня зміна за один крок становить 0.5, то через 3 одиниці часу середня зміна буде:

$E(X_3) = 3 \cdot E(\Delta X) = 3 \cdot 0.5 = 1.5$

Відповідь:
Середня ціна акції через 3 одиниці часу дорівнює 1.5.

````python
import numpy as np
import matplotlib.pyplot as plt

def stock_price_at_time(t):
    # Симуляція випадкового блукання для часу t
    price = 0
    for _ in range(t):
        price += np.random.choice([-1, 2])  # Збільшення на 2 або зменшення на 1
    return price

def simulate_n_times(n: int, t: int) -> list:
    np.random.seed(42)
    return [stock_price_at_time(t) for _ in range(n)]
````

````python
# Симуляція і побудова гістограми для різних кількостей симуляцій
t = 3  # Час, для якого ми хочемо симулювати

for n in [10, 100, 1000, 10000]:
    prices = simulate_n_times(n, t)

    # Виводимо середнє значення
    mean_price = np.mean(prices)
    print(f"Середнє значення для {n} симуляцій: {mean_price}")

    # Побудова гістограми
    plt.figure(figsize=(8, 4))
    plt.hist(prices, bins=10, edgecolor='black')
    plt.title(f'Гістограма ціни через {t} одиниць часу для {n} симуляцій')
    plt.xlabel('Ціна акцій')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()
````
