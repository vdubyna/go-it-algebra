````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normaltest

# Параметри гамма-розподілу
shape_param = 0.3  # параметр форми
scale_param = 1.1  # параметр масштабу

# Симуляція зміни ціни для часу t з гамма-розподілом
def stock_price_at_time_gamma(t):
    price = 0
    for _ in range(t):
        # Зміна ціни акцій з гамма-розподілу
        price += np.random.gamma(shape_param, scale_param)
    return price

# Функція для симуляції n разів для часу t
def simulate_n_times_gamma(n: int, t: int) -> list:
    np.random.seed(42)
    return [stock_price_at_time_gamma(t) for _ in range(n)]

# Побудова гістограми та перевірка на нормальність
def plot_histogram_and_check_normality(prices, t, n):
    plt.figure(figsize=(8, 4))
    plt.hist(prices, bins=10, edgecolor='black', density=True)
    plt.title(f'Гістограма ціни через {t} одиниць часу для {n} симуляцій')
    plt.xlabel('Ціна акцій')
    plt.ylabel('Щільність')
    plt.grid(True)
    plt.show()

    # Тест на нормальність
    stat, p_value = normaltest(prices)
    print(f"Тест на нормальність для t={t}: Статистика = {stat}, p-значення = {p_value}")
    if p_value > 0.05:
        print(f"Для t={t} розподіл можна вважати нормальним.")
    else:
        print(f"Для t={t} розподіл не є нормальним.")

# Основна частина
n = 100  # кількість симуляцій
t_values = range(1, 61, 2)  # t від 1 до 60 з кроком 2

for t in t_values:
    prices = simulate_n_times_gamma(n, t)

    # Побудова гістограми та перевірка на нормальність
    plot_histogram_and_check_normality(prices, t, n)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Завантаження даних
url = 'https://docs.google.com/spreadsheets/d/18WCpPS96Tb3cB0FCsIA92PEhcmBkp08sjYhS9DsQfJE/edit#gid=954244094'
url = url[:url.find('/edit')] + '/export?format=csv'
data = pd.read_csv(url)

# Функція для обчислення основних статистичних показників
def calculate_statistics(column):
    mean = np.mean(column)
    variance = np.var(column)
    std_dev = np.std(column)
    return mean, variance, std_dev

# Перевірка на нормальність розподілу (тест Шапіро-Вілка)
def check_normality(column):
    stat, p_value = stats.shapiro(column)
    if p_value > 0.05:
        return 'Розподіл нормальний'
    else:
        return 'Розподіл не є нормальним'

# Підрахунок кореляції
def calculate_correlation(column, target_column):
    correlation = np.corrcoef(column, target_column)[0, 1]
    return correlation

# Збір результатів у DataFrame
results = pd.DataFrame(columns=["Показник", "Середнє значення", "Дисперсія", "Стандартне відхилення", "Нормальність", "Кореляція з Product_Sold"])

# Цикл по кожній колонці
for i, column in enumerate(data.columns):  # Остання колонка - Product_Sold
    mean, variance, std_dev = calculate_statistics(data[column])
    normality = check_normality(data[column])
    correlation = calculate_correlation(data[column], data['Product_Sold'])

    # Додавання результатів до DataFrame за допомогою .loc[]
    results.loc[i] = [column, mean, variance, std_dev, normality, correlation]

# Показуємо результати у вигляді таблиці
print("\nРезультати аналізу витрат на рекламу:\n")
print(results)

# Побудова гістограм для кожного показника
for column in data.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[column], bins=20, kde=True, edgecolor='black')
    plt.title(f'Гістограма розподілу для {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()
````
