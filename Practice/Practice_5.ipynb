# Simplex method

[explanation](https://youtu.be/E72DWgKP_1Y?feature=shared)
[proof](https://www.cse.iitb.ac.in/~cs435/scribe/lecture13a.pdf)

[example](https://math.libretexts.org/Bookshelves/Applied_Mathematics/Applied_Finite_Mathematics_(Sekhon_and_Bloom)/04%3A_Linear_Programming_The_Simplex_Method/4.02%3A_Maximization_By_The_Simplex_Method)

$
\begin{cases}
Z = 40x_1+30x_2\\
x_1 + x_2 \le 12\\
2x_1 + x_2 \le 16\\
x_1, x_2 \ge 0
\end{cases}\\
\begin{cases}
Z = 40x_1+30x_2\\
x_1 + x_2 + y_1 = 12\\
2x_1 + x_2 + y_2 = 16\\
x_1, x_2 \ge 0
\end{cases}\\
\begin{cases}
Z = 40x_1+30x_2\\
y_1 = 12 - x_1 - x_2\\
y_2 = 16 - 2x_1 - x_2\\
x_1, x_2 \ge 0
\end{cases}
$

$
\textbf{обираєм максимальний коефіцієнт цільової функції:}\\
\begin{cases}
Z = 40x_1+30x_2 | 40 - \max\\
y_1 = 12 - x_1 - x_2\\
y_2 = 16 - 2x_1 - x_2\\
x_1, x_2 \ge 0
\end{cases}\\
\textbf{співвідношення константи до коеф. біля } x_1:\\
\begin{cases}
\frac{12}{-1} = -12\\
\frac{16}{-2} = -8 - \textbf{маскимальне, не позитивне співвідношення}\\
\end{cases}\\
\textbf{виражаєм } x_1 \textbf{з 2 рівняння:}\\
\begin{cases}
Z = 40x_1+30x_2\\
y_1 = 12 - x_1 - x_2\\
x_1 = 8 - \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1, x_2 \ge 0
\end{cases}\\
\textbf{підставляєм в 1 рівняння та цільову функцію:}\\
\begin{cases}
Z = 40(8 - \frac{1}{2}y_2 - \frac{1}{2}x_2) + 30x_2 = 320 - 20y_2 - 20x_2 + 30x_2 = 320 - 20y_2 + 10x_2\\
y_1 = 12 - (8 - \frac{1}{2}y_2 - \frac{1}{2}x_2) - x_2 = 4 + \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1 = 8 - \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1, x_2 \ge 0
\end{cases}\\
\textbf{обираєм максимальний коефіцієнт цільової функції:}\\
\begin{cases}
Z = 320 - 20y_2 + 10x_2 | 10 - \max\\
y_1 = 4 + \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1 = 8 - \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1, x_2 \ge 0
\end{cases}\\
$

$
\textbf{співвідношення константи до коеф. біля } x_2:\\
\begin{cases}
\frac{4}{-\frac{1}{2}} = -8 - \textbf{маскимальне, не позитивне співвідношення}\\
\frac{8}{-\frac{1}{2}} = -16\\
\end{cases}\\
\textbf{виражаєм } x_2 \textbf{ з 1 рівняння:}\\
\begin{cases}
Z = 320 - 20y_2 + 10x_2\\
x_2 = 8 + y_2 - 2y_1\\
x_1 = 8 - \frac{1}{2}y_2 - \frac{1}{2}x_2\\
x_1, x_2 \ge 0
\end{cases}\\
\textbf{підставляєм в 2 рівняння та цільову функцію:}\\
\begin{cases}
Z = 320 - 20y_2 + 10(8 + y_2 - 2y_1) = 320 - 20y_2 + 80 + 10y_2 - 20y_1 = 400 - 10y_2 - 20y_1\\
x_2 = 8 + y_2 - 2y_1\\
x_1 = 8 - \frac{1}{2}y_2 - \frac{1}{2}(8 + y_2 - 2y_1) = 8 - \frac{1}{2}y_2 - 4 - \frac{1}{2}y_2 + y_1 = 4 - y_2 + y_1\\
x_1, x_2 \ge 0
\end{cases}\\
\begin{cases}
Z = 400 - 10y_2 - 20y_1\\
x_2 = 8 + y_2 - 2y_1\\
x_1 = 4 - y_2 + y_1\\
x_1, x_2 \ge 0
\end{cases}\\
\textbf{більше позитивних коефіцієнтів в цільовій функції немає, тож можемо записати відповідь: }\\
\begin{cases}
Z = 400\\
x_2 = 8\\
x_1 = 4\\
\end{cases}\\
$

```python
np.random.seed(42)
n = 500
x = np.random.uniform(4, 8, n)
f = 0.5*x + np.cos(10*x) + 2*np.sin(3*x) + 0.01*(x-5)**5 + 5 + np.random.normal(0, 0.3, n)
df = pd.DataFrame({'x': x, 'f': f})
df
```

```python
# df['x2'] = np.sqrt(df.x)
```

```python
sns.scatterplot(x=df.x, y=df.f)
```

```python
from sklearn.model_selection import train_test_split
target = 'f'
X = df.drop(columns=[target])
y = df[target]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
sns.scatterplot(x=x_train.x, y=y_train, color='red', alpha=0.6, label='train')
sns.scatterplot(x=x_test.x, y=y_test, color='green', alpha=0.6, label='test')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score, get_scorer_names
```
